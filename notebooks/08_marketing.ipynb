````
# ⚙️ 환경 설정 (처음 1회 실행)
!pip install -q -U google-genai

from google import genai
from google.colab import userdata

# 🔑 보안 비밀에서 API 키 로드
API_KEY = userdata.get('GEMINI_API_KEY')

client = genai.Client(api_key=API_KEY)
MODEL = 'gemini-2.5-flash'
print('✅ Gemini API 연결 완료!')
````

````
# ✏️ 학원 정보
academy_name = "달빛영어"
academy_location = "서울시 강남구"
campaign_topic = """여름 방학 특강 모집
대상: 초등 3~6학년
기간: 7월 22일 ~ 8월 16일 (4주)
특징: 원어민 스피킹 집중 + 파닉스 완성
혜택: 얼리버드 20% 할인 (6/30까지)"""
````

## 🔬 실습 1: AIDA 블로그 포스트

````
prompt1 = f"""
너는 교육 분야 블로그 마케팅 전문가야.

AIDA 프레임워크로 블로그 포스트를 작성해줘.

학원: {academy_name} ({academy_location})
캠페인: {campaign_topic}

AIDA 구조:
A (Attention) — 클릭할 수밖에 없는 제목 + 도입부
I (Interest) — 학부모의 고민 공감 + 데이터
D (Desire) — 우리 프로그램의 차별점 + 후기
A (Action) — 명확한 CTA (전화번호, 링크)

SEO 키워드: 강남 영어학원, 초등 영어, 여름 방학 영어
1500자 이상. 네이버 블로그 스타일.
"""

response = client.models.generate_content(model=MODEL, contents=prompt1)
blog = response.text
print('📝 AIDA 블로그 포스트')
print('='*50)
print(blog)
````

## 🔬 실습 2: 5채널 마케팅 카피

````
channels = [
    ('인스타그램', '이모지 활용. 해시태그 10개. 150자 이내. 비주얼 중심.'),
    ('카카오톡 단체 메시지', '학부모 대상. 100자 이내. 링크 포함. 친근한 톤.'),
    ('네이버 밴드', '동네 맘 커뮤니티 스타일. 200자. 질문형 시작.'),
    ('당근마켓', '동네 타겟. 가격 정보 포함. 150자. 신뢰감 있는 톤.'),
    ('전단지/포스터', '한눈에 들어오는 레이아웃. 핵심 정보만. 50자 이내 헤드라인.')
]

all_copy = []
for channel, style in channels:
    prompt = f"""
너는 {channel} 마케팅 카피라이터야.

아래 캠페인의 {channel}용 카피를 작성해줘.

학원: {academy_name}
캠페인: {campaign_topic}
스타일: {style}
"""
    response = client.models.generate_content(model=MODEL, contents=prompt)
    all_copy.append(response.text)
    print(f'\n📱 {channel}')
    print('='*50)
    print(response.text)
````

## 🔬 실습 3: 월간 콘텐츠 캘린더

````
prompt3 = f"""
너는 학원 SNS 마케팅 매니저야.

{academy_name}의 1개월 콘텐츠 캘린더를 만들어줘.

| 주차 | 월 | 화 | 수 | 목 | 금 |
|---|---|---|---|---|---|
| 1주 | 콘텐츠 | 콘텐츠 | 콘텐츠 | 콘텐츠 | 콘텐츠 |

콘텐츠 유형: 수업 후기, 학생 활동, 학습 팁, 이벤트, 학부모 후기
채널: 인스타 / 블로그 / 밴드 번갈아 사용
각 칸에: [채널] 제목 (형식)
주 3회 발행 기준 (월수금).
"""

response = client.models.generate_content(model=MODEL, contents=prompt3)
calendar = response.text
print('📅 월간 콘텐츠 캘린더')
print('='*50)
print(calendar)
````

````
# 💾 저장
output = f"블로그:\n{blog}\n\n캘린더:\n{calendar}"
for i, (ch, _) in enumerate(channels):
    output += f"\n\n{ch}:\n{all_copy[i]}"
with open('마케팅_패키지.txt', 'w', encoding='utf-8') as f:
    f.write(output)
from google.colab import files
files.download('마케팅_패키지.txt')
print('✅ 다운로드!')
````
