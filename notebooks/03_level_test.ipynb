````
# ⚙️ 환경 설정 (처음 1회 실행)
!pip install -q -U google-genai

from google import genai
from google.colab import userdata

# 🔑 보안 비밀에서 API 키 로드
API_KEY = userdata.get('GEMINI_API_KEY')

client = genai.Client(api_key=API_KEY)
MODEL = 'gemini-2.5-flash'
print('✅ Gemini API 연결 완료!')
````

````
# ✏️ 설정
academy_name = "달빛영어"
test_target = "초등 3~6학년 신규 상담생"
````

## 🔬 실습 1: Pre-A1 레벨테스트 (20문항)

````
prompt1 = """
너는 CEFR 기반 영어 평가 전문가야.

Pre-A1 레벨 학생용 레벨테스트 20문항을 만들어줘.

구성:
- 파트 1: 그림 보고 단어 고르기 (5문항) — 알파벳, 색깔, 숫자
- 파트 2: 듣고 따라 말하기 (5문항) — 인사, 자기소개
- 파트 3: 매칭 (5문항) — 그림과 단어 연결
- 파트 4: 빈칸 채우기 (5문항) — 기초 어휘

각 문항: 번호 | 문제 | 보기(4지선다) | 정답 | 평가영역
초등 저학년이 이해할 수 있는 쉬운 영어로.
Google Forms로 바로 옮길 수 있는 형식.
"""

response = client.models.generate_content(model=MODEL, contents=prompt1)
pre_a1 = response.text
print('📝 Pre-A1 레벨테스트')
print('='*50)
print(pre_a1)
````

## 🔬 실습 2: A1 레벨테스트 (20문항)

````
prompt2 = """
너는 CEFR 기반 영어 평가 전문가야.

A1 레벨 학생용 레벨테스트 20문항을 만들어줘.

구성:
- 파트 1: 어휘 (5문항) — 일상 어휘, 반의어
- 파트 2: 문법 (5문항) — be동사, 인칭대명사, 현재시제
- 파트 3: 읽기 이해 (5문항) — 짧은 지문 (3~5문장)
- 파트 4: 쓰기 (5문항) — 문장 완성, 어순 배열

각 문항: 번호 | 문제 | 보기(4지선다) | 정답 | 평가영역
"""

response = client.models.generate_content(model=MODEL, contents=prompt2)
a1 = response.text
print('📝 A1 레벨테스트')
print('='*50)
print(a1)
````

## 🔬 실습 3: A2 레벨테스트 (20문항)

````
prompt3 = """
너는 CEFR 기반 영어 평가 전문가야.

A2 레벨 학생용 레벨테스트 20문항을 만들어줘.

구성:
- 파트 1: 어휘 (5문항) — 주제별 어휘 (학교, 취미, 가족)
- 파트 2: 문법 (5문항) — 과거시제, 비교급, 조동사
- 파트 3: 읽기 이해 (5문항) — 중간 길이 지문 (5~8문장)
- 파트 4: 쓰기 (5문항) — 자유 문장 작성, 이메일 작성

각 문항: 번호 | 문제 | 보기(4지선다) | 정답 | 평가영역
"""

response = client.models.generate_content(model=MODEL, contents=prompt3)
a2 = response.text
print('📝 A2 레벨테스트')
print('='*50)
print(a2)
````

## 🔬 실습 4: 학부모 결과 안내문

````
prompt4 = f"""
너는 {academy_name}의 상담 매니저야.

레벨테스트 결과를 학부모에게 안내하는 편지를 3종 작성해줘.
(Pre-A1 결과 / A1 결과 / A2 결과)

각 편지 포함 내용:
1. 인사말
2. 테스트 결과 요약 (강점 + 보완점)
3. 추천 반 배치
4. 학습 로드맵 (3개월)
5. 다음 단계 안내

학부모가 안심할 수 있는 따뜻한 톤으로.
각 편지 300자 이내.
"""

response = client.models.generate_content(model=MODEL, contents=prompt4)
letters = response.text
print('💌 학부모 안내문 3종')
print('='*50)
print(letters)
````

````
# 💾 저장
output = f"Pre-A1:\n{pre_a1}\n\nA1:\n{a1}\n\nA2:\n{a2}\n\n안내문:\n{letters}"
with open('레벨테스트_패키지.txt', 'w', encoding='utf-8') as f:
    f.write(output)
from google.colab import files
files.download('레벨테스트_패키지.txt')
print('✅ 다운로드!')
````
