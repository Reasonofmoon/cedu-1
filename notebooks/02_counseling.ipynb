````
# ⚙️ 환경 설정 (처음 1회 실행)
!pip install -q -U google-genai

from google import genai
from google.colab import userdata

# 🔑 보안 비밀에서 API 키 로드
API_KEY = userdata.get('GEMINI_API_KEY')

client = genai.Client(api_key=API_KEY)
MODEL = 'gemini-2.5-flash'
print('✅ Gemini API 연결 완료!')
````

````
# ✏️ 학원 정보
academy_name = "달빛영어"
student_name = "김서연"
student_info = """초5 / B반 / 재원 8개월
최근 성적: 읽기 85 → 78 (하락)
강점: 어휘력 우수
약점: 문법 기초 부족"""
````

## 🔬 실습 1: 4유형 학부모 브리핑 스크립트

````
parent_types = ['성과중시형', '불안형', '방임형', '과잉개입형']

for p_type in parent_types:
    prompt = f"""
너는 학원 상담 전문가 10년 경력이야.

'{p_type}' 유형의 학부모에게 아래 학생 상황을 브리핑하는 스크립트를 작성해줘.

학생: {student_name}
정보: {student_info}

스크립트 구성:
1. 인사 & 아이스브레이킹 (1문장)
2. 긍정적 시작 — 칭찬 포인트 (2문장)
3. 현황 보고 — 데이터 기반 (3문장)
4. 개선 방안 — 구체적 제안 (3문장)
5. 마무리 — 약속 & 다음 상담 예고 (2문장)

'{p_type}' 학부모의 심리적 특성에 맞춰 톤과 강조점을 조절해줘.
500자 이내로.
"""
    response = client.models.generate_content(model=MODEL, contents=prompt)
    print(f'\n👤 {p_type} 학부모 브리핑')
    print('='*50)
    print(response.text)
    print()
````

## 🔬 실습 2: 반박 대응 스크립트

````
prompt2 = f"""
너는 학원 상담 위기관리 전문가야.

학부모가 자주 하는 불만/반박 5가지에 대한 대응 스크립트를 만들어줘.

| 번호 | 학부모 불만 | 공감 표현 | 사실 기반 반박 | 대안 제시 |
|---|---|---|---|---|
| 1 | "성적이 안 올라요" | ... | ... | ... |
| 2 | "숙제가 너무 많아요" | ... | ... | ... |
| 3 | "다른 학원은 더 싸요" | ... | ... | ... |
| 4 | "아이가 재미없대요" | ... | ... | ... |
| 5 | "효과가 있는 건가요?" | ... | ... | ... |

학원: {academy_name}
공감 → 팩트 → 대안 순서로.
"""

response = client.models.generate_content(model=MODEL, contents=prompt2)
rebuttals = response.text
print('🛡️ 반박 대응 스크립트')
print('='*50)
print(rebuttals)
````

## 🔬 실습 3: 상담 Gem 시스템 프롬프트

````
prompt3 = f"""
너는 AI 시스템 설계자이자 학원 상담 코치야.

학부모 상담을 도와주는 Gemini Gem의 시스템 프롬프트를 작성해줘.

학원: {academy_name}

Gem이 해야 할 일:
- 학생 정보를 입력받으면 상담 브리핑 스크립트 생성
- 학부모 유형에 맞는 톤 조절
- 반박 상황에서 대응 스크립트 제공
- 상담 후 follow-up 메시지 생성

Gem 프롬프트 형식:
---
이름: {academy_name} 상담 도우미
역할: ...
지침: ...
출력 형식: ...
제약: ...
---

실제 학원 업무에 바로 사용할 수 있게.
"""

response = client.models.generate_content(model=MODEL, contents=prompt3)
gem_prompt = response.text
print('🤖 상담 Gem 시스템 프롬프트')
print('='*50)
print(gem_prompt)
print('\n💡 이 프롬프트를 Gemini → Gem Manager → New Gem에 붙여넣으세요!')
````

````
# 💾 저장
output = f"반박 대응:\n{rebuttals}\n\nGem 프롬프트:\n{gem_prompt}"
with open('상담_브리핑_패키지.txt', 'w', encoding='utf-8') as f:
    f.write(output)
from google.colab import files
files.download('상담_브리핑_패키지.txt')
print('✅ 다운로드!')
````
