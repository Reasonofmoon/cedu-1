````
# ⚙️ 환경 설정 (처음 1회 실행)
!pip install -q -U google-genai

from google import genai
from google.colab import userdata

# 🔑 보안 비밀에서 API 키 로드
API_KEY = userdata.get('GEMINI_API_KEY')

client = genai.Client(api_key=API_KEY)
MODEL = 'gemini-2.5-flash'
print('✅ Gemini API 연결 완료!')
````

````
# ✏️ 학원 정보
academy_name = "달빛영어"
````

## 🔬 실습 1: 5유형 이탈 분류 & 맞춤 메시지

````
churn_types = [
    ('성적 불만형', '성적이 오르지 않아 퇴원을 고려하는 학부모'),
    ('경제적 부담형', '수업료 부담으로 퇴원을 고려하는 학부모'),
    ('이사/전학형', '이사나 전학으로 물리적으로 다닐 수 없는 학부모'),
    ('경쟁 학원 이탈형', '다른 학원의 마케팅에 끌려 이동하려는 학부모'),
    ('관심 저하형', '영어 학습 자체에 대한 관심이 줄어든 학부모')
]

all_messages = []
for ctype, desc in churn_types:
    prompt = f"""
너는 학원 재등록 상담 전문가 15년 경력이야.

'{ctype}' ({desc})에게 보내는 메시지 3종을 작성해줘.

학원: {academy_name}

1. 📱 카카오톡 메시지 (100자, 따뜻한 톤)
2. 📞 전화 스크립트 (200자, 공감 → 해결책)
3. 💌 재등록 권유 편지 (300자, 데이터 + 혜택)

각 메시지에 구체적인 혜택이나 대안을 포함해줘.
"""
    response = client.models.generate_content(model=MODEL, contents=prompt)
    all_messages.append(response.text)
    print(f'\n📌 {ctype}')
    print('='*50)
    print(response.text)
````

## 🔬 실습 2: 30일 리텐션 캠페인

````
prompt2 = f"""
너는 학원 마케팅 & CRM 전문가야.

퇴원 의사를 밝힌 학부모를 대상으로 30일 리텐션 캠페인을 설계해줘.

학원: {academy_name}

| 일차 | 채널 | 액션 | 메시지 핵심 | 담당 |
|---|---|---|---|---|
| D+1 | 카톡 | ... | ... | 원장 |
| D+3 | 전화 | ... | ... | 담임 |
| ... | ... | ... | ... | ... |
| D+30 | 방문 | ... | ... | 원장 |

30일을 10개 터치포인트로 나눠줘.
앞부분: 감정 케어 / 중간: 데이터 제시 / 후반: 혜택 제안.
"""

response = client.models.generate_content(model=MODEL, contents=prompt2)
campaign = response.text
print('📅 30일 리텐션 캠페인')
print('='*50)
print(campaign)
````

## 🔬 실습 3: 재등록 혜택 패키지

````
prompt3 = f"""
너는 학원 경영 전략가야.

재등록을 유도하는 혜택 패키지 3종을 설계해줘.

학원: {academy_name}

패키지 A: 💰 경제적 혜택 (할인, 무료 수업 등)
패키지 B: 📚 학습 혜택 (추가 수업, 1:1 코칭 등)
패키지 C: 🎁 프리미엄 혜택 (A+B 결합)

각 패키지:
- 이름과 슬로건
- 포함 혜택 5가지
- 유효 기간
- 예상 비용 vs 재등록 수익
- 학부모 안내 멘트

실현 가능하고 학원 수익에 도움 되는 구조로.
"""

response = client.models.generate_content(model=MODEL, contents=prompt3)
packages = response.text
print('🎁 재등록 혜택 패키지')
print('='*50)
print(packages)
````

````
# 💾 저장
output = f"캠페인:\n{campaign}\n\n혜택 패키지:\n{packages}"
with open('재등록_캠페인.txt', 'w', encoding='utf-8') as f:
    f.write(output)
from google.colab import files
files.download('재등록_캠페인.txt')
print('✅ 다운로드!')
````
