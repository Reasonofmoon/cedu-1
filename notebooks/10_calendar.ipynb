````
# ⚙️ 환경 설정 (처음 1회 실행)
!pip install -q -U google-genai

from google import genai
from google.colab import userdata

# 🔑 보안 비밀에서 API 키 로드
API_KEY = userdata.get('GEMINI_API_KEY')

client = genai.Client(api_key=API_KEY)
MODEL = 'gemini-2.5-flash'
print('✅ Gemini API 연결 완료!')
````

````
# ✏️ 학원 정보
academy_name = "달빛영어"
current_students = 45
target_students = 70
````

## 🔬 실습 1: 4시즌 전략표

````
prompt1 = f"""
너는 학원 경영 전략 컨설턴트 20년 경력이야.

{academy_name}의 4시즌 연간 전략표를 만들어줘.

현재: {current_students}명 / 목표: {target_students}명

각 시즌별:
| 시즌 | 기간 | 핵심 목표 | 전략 3가지 | 핵심 이벤트 2개 | 주의점 | KPI 2개 |

🌸 봄 (3-5월) — ACQUIRE: 신규 모집
☀️ 여름 (6-8월) — INTENSIFY: 방학 특강
🍂 가을 (9-11월) — RETAIN: 유지 관리
❄️ 겨울 (12-2월) — PLAN: 결산 & 설계

개인 학원 기준. ROI 높은 전략 위주.
"""

response = client.models.generate_content(model=MODEL, contents=prompt1)
season = response.text
print('🌸☀️🍂❄️ 4시즌 전략표')
print('='*50)
print(season)
````

## 🔬 실습 2: 12개월 운영 계획

````
prompt2 = f"""
너는 학원 운영 매니저 전문가야.

{academy_name}의 2026년 1월~12월 월별 운영 계획표를 만들어줘.

| 월 | 시즌 | 핵심 이벤트 | 마케팅 | 학사 | 재무 | 슬로건 |
|---|---|---|---|---|---|---|

한국 학교 일정(학기, 방학, 시험)에 맞춰줘.
2026년 기준.
"""

response = client.models.generate_content(model=MODEL, contents=prompt2)
monthly = response.text
print('📅 12개월 운영 계획')
print('='*50)
print(monthly)
````

## 🔬 실습 3: KPI 10개 + Sheets 수식

````
prompt3 = f"""
너는 학원 경영 데이터 분석 전문가야.

학원 운영의 핵심 KPI 10개를 설정하고 Google Sheets에서 추적할 수 있는 구조를 만들어줘.

KPI 10개:
1. 월 신규 등록 수
2. 월 퇴원 수
3. 재원생 유지율
4. 평균 재원 기간
5. 레벨 상승 비율
6. 학부모 만족도
7. 수업 출석률
8. 숙제 완료율
9. 레벨테스트 평균
10. 마케팅 전환율

각 KPI: 목표값 | 현재값 | 달성률 수식 | 상태(🟢🟡🔴)
Google Sheets 수식 예시도 정확하게.
매주 10분이면 업데이트 가능한 구조.
"""

response = client.models.generate_content(model=MODEL, contents=prompt3)
kpi = response.text
print('📊 KPI 10개 시스템')
print('='*50)
print(kpi)
````

## 🔬 실습 4: 주간 루틴 템플릿

````
prompt4 = """
너는 학원장 시간 관리 코치야.

1인 원장의 주간 루틴 템플릿을 만들어줘.

| 요일 | 오전 | 수업시간 | 오후 | 관리업무 |
|---|---|---|---|---|

포함 업무: 상담(주2-3), 강사미팅(주1), 마케팅(주1), KPI(주1),
학부모소통(매일), 수업관찰(주1), 재무점검(주1)

관리 업무는 하루 2시간 이내. 주말 최소화.
"""

response = client.models.generate_content(model=MODEL, contents=prompt4)
routine = response.text
print('📋 주간 루틴')
print('='*50)
print(routine)
````

````
# 💾 저장
output = f"4시즌:\n{season}\n\n월별:\n{monthly}\n\nKPI:\n{kpi}\n\n루틴:\n{routine}"
with open('연간운영_캘린더.txt', 'w', encoding='utf-8') as f:
    f.write(output)
from google.colab import files
files.download('연간운영_캘린더.txt')
print('✅ 다운로드!')
````
