````
# ⚙️ 환경 설정 (처음 1회 실행)
!pip install -q -U google-genai

from google import genai
from google.colab import userdata

# 🔑 보안 비밀에서 API 키 로드
API_KEY = userdata.get('GEMINI_API_KEY')

client = genai.Client(api_key=API_KEY)
MODEL = 'gemini-2.5-flash'
print('✅ Gemini API 연결 완료!')
````

````
# ✏️ 우리 학원 정보 (여기를 수정하세요!)
academy_name = "달빛영어"          # ← 학원 이름
academy_location = "서울시 강남구"   # ← 위치
academy_target = "초등 3~6학년"     # ← 대상
academy_concept = """파닉스부터 스피킹까지,
놀이 중심의 영어 몰입 교육"""        # ← 교육 컨셉 (여러 줄 가능)

print(f'🏫 {academy_name} 정보 설정 완료!')
````

## 🔬 실습 1: 브랜드 브리프 생성

````
prompt1 = f"""
너는 교육 브랜딩 전문 컨설턴트 10년 경력이야.

아래 정보로 우리 학원의 브랜드 브리프(Brand Brief)를 작성해줘.

학원명: {academy_name}
위치: {academy_location}
대상: {academy_target}
컨셉: {academy_concept}

브랜드 브리프 형식:
1. 미션 (Mission) — 1문장
2. 비전 (Vision) — 1문장
3. 핵심 가치 (Core Values) — 3개
4. 타겟 고객 프로필 — 표 형식
5. 브랜드 톤앤매너 — 3가지 키워드
6. 경쟁 차별점 (USP) — 3가지
7. 브랜드 스토리 — 100자 이내

학원 원장이 바로 사용할 수 있는 실용적 문서로.
"""

response = client.models.generate_content(model=MODEL, contents=prompt1)
brand_brief = response.text
print('📋 브랜드 브리프')
print('='*50)
print(brand_brief)
````

## 🔬 실습 2: 슬로건 5종 생성

````
prompt2 = f"""
너는 카피라이터 15년 경력이야.

{academy_name}의 브랜드 슬로건 5개를 만들어줘.

브랜드 브리프:
{brand_brief}

| 번호 | 슬로건 | 컨셉 | 활용처 |
|---|---|---|---|
| 1 | (10자 이내) | ... | 간판/명함 |
| 2 | ... | ... | 전단지 |
| 3 | ... | ... | SNS |
| 4 | ... | ... | 학부모 상담 |
| 5 | ... | ... | 캐치프레이즈 |

과장하지 말고 실현 가능한 약속만.
"""

response = client.models.generate_content(model=MODEL, contents=prompt2)
slogans = response.text
print('✨ 슬로건 5종')
print('='*50)
print(slogans)
````

## 🔬 실습 3: 학부모 FAQ 10개

````
prompt3 = f"""
너는 학원 상담 매니저 7년 경력이야.

{academy_name}에 대해 학부모가 가장 자주 묻는 질문(FAQ) 10개와 모범 답변을 만들어줘.

형식:
### Q1. [질문]
**A.** [답변 — 3문장 이내]

주제: 수업료, 반편성, 숙제, 효과, 수업방식, 교재, 레벨테스트, 시간표, 보강, 상담
학부모 눈높이에 맞는 친근한 톤으로.
"""

response = client.models.generate_content(model=MODEL, contents=prompt3)
faq = response.text
print('❓ FAQ 10개')
print('='*50)
print(faq)
````

## 🔬 실습 4: 상담용 A4 1장

````
prompt4 = f"""
너는 학원 마케팅 디자이너이자 상담 전문가야.

{academy_name}의 상담용 A4 한 장짜리 안내문을 만들어줘.

브랜드 브리프:
{brand_brief}

구성:
📌 상단: 학원명 + 슬로건
📌 중단: 교육 특징 3가지 (아이콘 스타일)
📌 하단: 수업 시간표 (표) + 연락처

Markdown 형식으로 작성.
인쇄해서 바로 사용할 수 있게.
"""

response = client.models.generate_content(model=MODEL, contents=prompt4)
a4_sheet = response.text
print('📄 상담용 A4')
print('='*50)
print(a4_sheet)
````

````
# 💾 전체 결과 저장 & 다운로드
output = f"""{'='*60}
🎨 {academy_name} 브랜딩 패키지
{'='*60}

📋 1. 브랜드 브리프
{'-'*40}
{brand_brief}

✨ 2. 슬로건 5종
{'-'*40}
{slogans}

❓ 3. FAQ 10개
{'-'*40}
{faq}

📄 4. 상담용 A4
{'-'*40}
{a4_sheet}
"""

with open('브랜딩_패키지.txt', 'w', encoding='utf-8') as f:
    f.write(output)

from google.colab import files
files.download('브랜딩_패키지.txt')
print('✅ 다운로드 완료!')
````
