*(왼쪽 사이드바 🔑 아이콘 → 새 보안 비밀 추가)*

````
# ⚙️ 환경 설정 (처음 1회 실행)
!pip install -q -U google-genai

from google import genai
from google.colab import userdata

# 🔑 보안 비밀에서 API 키 로드
API_KEY = userdata.get('GEMINI_API_KEY')

client = genai.Client(api_key=API_KEY)
MODEL = 'gemini-2.5-flash'
print('✅ Gemini API 연결 완료!')
````

````
# ✏️ 학원 정보 (여러 줄 붙여넣기 가능!)
academy_name = "달빛영어"
brand_brief = """여기에 1강에서 만든 브랜드 브리프를 붙여넣으세요.
여러 줄로 붙여넣어도 됩니다.
삼중 따옴표 안에 자유롭게 작성하세요."""
````

## 🔬 실습 1: 학원 운영 흐름도 (Mermaid)

````
prompt1 = f"""
너는 비즈니스 프로세스 설계 전문가야.

{academy_name}의 전체 운영 사이클을 Mermaid 다이어그램으로 만들어줘.
12강에서 배운 모든 것을 하나의 흐름으로 연결해.

흐름:
1. 브랜딩(1강) → 2. 상담(2강) → 3. 레벨테스트(3강)
→ 등록 → 4. 피드백(4강) → 5. 시험대비(5강)
→ 6. 성적분석(6강) → 위험감지?
→ YES → 7. 재등록(7강)
→ NO → 계속 → 8. 마케팅(8강) → 1번 순환

보조 프로세스:
→ 9. 수업표준화(9강) — 매 수업
→ 10. 캘린더(10강) — 매 분기
→ 11. 역량진단(11강) — 매 분기
→ 12. 대시보드(12강) — 항상

Mermaid graph LR 형식으로 작성.
각 노드에 이모지 + 강의번호 포함.
"""

response = client.models.generate_content(model=MODEL, contents=prompt1)
flowchart = response.text
print('🔄 운영 흐름도 (Mermaid)')
print('='*50)
print(flowchart)
print('\n💡 이 코드를 https://mermaid.live/ 에 붙여넣으면 다이어그램이 생성됩니다!')
````

## 🔬 실습 2: 전략 Gem 시스템 프롬프트

````
prompt2 = f"""
너는 AI 시스템 설계자이자 학원 경영 코치야.

학원 경영 전략을 상담해주는 Gemini Gem의 시스템 프롬프트를 작성해줘.

기능:
- 학원 경영 상담 (신규 모집, 마케팅, 운영)
- KPI 분석 조언
- 위기 상황 대응
- 연간 계획 수립
- 강사 관리

학원: {academy_name}
정보: {brand_brief}

Gem 프롬프트 형식:
---
이름: {academy_name} 경영 파트너
역할: ...
지침: ...
출력 형식: ...
제약: ...
---

데이터 기반 조언. 비현실적 제안 금지.
"""

response = client.models.generate_content(model=MODEL, contents=prompt2)
strategy_gem = response.text
print('🤖 전략 Gem 프롬프트')
print('='*50)
print(strategy_gem)
print('\n💡 Gemini → Gem Manager → New Gem에 붙여넣기!')
````

## 🔬 실습 3: KPI 대시보드 구조 설계

````
prompt3 = f"""
너는 학원 경영 대시보드 설계 전문가야.

{academy_name}의 통합 KPI 대시보드 Google Sheets 구조를 설계해줘.

구성:
1. 📊 KPI 게이지 5종 (목표/현재/달성률/상태)
   - 신규등록률, 유지율, 출석률, 성적향상률, 학부모만족도
2. 📈 월간 추이 차트 영역 설명
3. 🚨 위험 알림 규칙 (조건부 서식)
4. ✅ 이번 달 할 일 체크리스트

모든 수식을 정확한 셀 참조로 제공.
수동 입력 최소화. 매주 5분 업데이트 가능.
"""

response = client.models.generate_content(model=MODEL, contents=prompt3)
dashboard = response.text
print('📊 통합 대시보드 구조')
print('='*50)
print(dashboard)
````

## 🔬 실습 4: 12강 마스터 체크리스트

````
# 🏆 최종 체크리스트
checklist = """
╔══════════════════════════════════════════════════╗
║     🎓 CEDU 기초과정 12강 마스터 체크리스트       ║
╠══════════════════════════════════════════════════╣
║                                                  ║
║  1강: □ 브랜드 브리프  □ 슬로건  □ FAQ           ║
║  2강: □ 상담 Gem 설정  □ 4유형 브리핑              ║
║  3강: □ 레벨테스트 3종  □ Forms 퀴즈              ║
║  4강: □ 피드백 4종     □ 배치 처리                ║
║  5강: □ 시험대비 5종   □ 단어카드                  ║
║  6강: □ 성적 대시보드  □ 위험 학생 감지            ║
║  7강: □ 30일 캠페인    □ 혜택 패키지               ║
║  8강: □ AIDA 블로그    □ 5채널 카피                ║
║  9강: □ 수업계획서     □ 강사 매뉴얼               ║
║ 10강: □ 4시즌 전략     □ KPI 10개                  ║
║ 11강: □ 역량 진단 30문  □ 레이더 차트              ║
║ 12강: □ 통합 대시보드  □ 전략 Gem                  ║
║                                                  ║
╠══════════════════════════════════════════════════╣
║         🎉 모두 체크하면 기초과정 이수!           ║
╚══════════════════════════════════════════════════╝
"""
print(checklist)
````

````
# 💾 전체 결과 저장
output = f"""
{'='*60}
🏠 {academy_name} 통합 운영 가이드
{'='*60}

📊 1. 운영 흐름도 (Mermaid)
{'-'*40}
{flowchart}

🤖 2. 전략 Gem 프롬프트
{'-'*40}
{strategy_gem}

📋 3. 대시보드 구조
{'-'*40}
{dashboard}

{checklist}
"""

filename = f'{academy_name}_통합운영가이드.txt'
with open(filename, 'w', encoding='utf-8') as f:
    f.write(output)
from google.colab import files
files.download(filename)
print(f'✅ {filename} 다운로드!')
````
