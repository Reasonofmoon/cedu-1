````
# ⚙️ 환경 설정 (처음 1회 실행)
!pip install -q -U google-genai

from google import genai
from google.colab import userdata

# 🔑 보안 비밀에서 API 키 로드
API_KEY = userdata.get('GEMINI_API_KEY')

client = genai.Client(api_key=API_KEY)
MODEL = 'gemini-2.5-flash'
print('✅ Gemini API 연결 완료!')
````

````
# ✏️ 수업 정보
academy_name = "달빛영어"
lesson_target = "초5 B반"
lesson_unit = "Unit 7: Comparatives"
lesson_goal = "비교급 형용사 활용"
lesson_grammar = "taller than, more beautiful than"
lesson_vocab = "tall, short, heavy, light, beautiful, exciting"
````

## 🔬 실습 1: 50분 수업계획서

````
prompt1 = f"""
너는 영어교육과 교수이자 수업설계(Instructional Design) 전문가야.

50분 영어 수업의 표준 수업계획서를 작성해줘.

수업 정보:
- 대상: {lesson_target}
- 단원: {lesson_unit}
- 학습 목표: {lesson_goal}
- 핵심 문법: {lesson_grammar}
- 핵심 어휘: {lesson_vocab}

수업계획서 형식:
| 단계 | 시간 | 활동 | 교사 역할 | 학생 활동 | 자료 |
|---|---|---|---|---|---|
| Opening | 5분 | ... | ... | ... | ... |
| Input | 15분 | ... | ... | ... | ... |
| Practice | 20분 | ... | ... | ... | ... |
| Output | 5분 | ... | ... | ... | ... |
| Check | 5분 | ... | ... | ... | ... |

각 단계의 전환 신호(Transition Signal)도 포함해줘.
50분을 정확히 맞춰줘. Practice에서 학생 활동 2가지 이상.
Exit Ticket 문제 3개 포함.
"""

response = client.models.generate_content(model=MODEL, contents=prompt1)
lesson_plan = response.text
print('📋 50분 수업계획서')
print('='*50)
print(lesson_plan)
````

## 🔬 실습 2: 강사 매뉴얼

````
prompt2 = f"""
너는 학원 운영 매뉴얼 작성 전문가야.

신규 강사가 첫 날부터 수업할 수 있도록 단계별 강사 매뉴얼을 만들어줘.

매뉴얼 구성:
1. 📌 학원 교육 철학 (3줄)
2. 📅 일일 루틴 (시간표)
3. 📋 수업 전 준비 체크리스트 (5개)
4. 📖 수업 진행 가이드 (5단계 공식)
5. 📝 출석/성적 기록 방법
6. 💬 학부모 소통 규칙 (카톡 응답 가이드)
7. ⚠️ 비상 상황 대응 (학생 울음, 다툼 등)
8. 📊 주간/월간 보고서 작성법

학원: {academy_name}
신규 강사 기준으로 아주 상세하게.
각 항목은 구체적 예시 포함.
"""

response = client.models.generate_content(model=MODEL, contents=prompt2)
manual = response.text
print('📖 강사 매뉴얼')
print('='*50)
print(manual)
````

## 🔬 실습 3: 수업 체크리스트 (전/중/후)

````
prompt3 = """
너는 수업 품질 관리 전문가야.

영어 수업의 전/중/후 체크리스트를 만들어줘.

📋 수업 전 (5분 전 점검) — 10개 항목
📋 수업 중 (진행 중 점검) — 10개 항목
📋 수업 후 (끝난 직후 점검) — 10개 항목

인쇄해서 강의실에 붙일 수 있는 형태로.
□ 체크박스 형식으로.
"""

response = client.models.generate_content(model=MODEL, contents=prompt3)
checklist = response.text
print('✅ 수업 체크리스트')
print('='*50)
print(checklist)
````

````
# 💾 저장
output = f"수업계획서:\n{lesson_plan}\n\n강사매뉴얼:\n{manual}\n\n체크리스트:\n{checklist}"
with open('수업표준화_매뉴얼.txt', 'w', encoding='utf-8') as f:
    f.write(output)
from google.colab import files
files.download('수업표준화_매뉴얼.txt')
print('✅ 다운로드!')
````
